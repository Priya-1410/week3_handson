```python
%%writefile run_all.sh
#!/usr/bin/env bash

# PrimoAgent Reproducibility Script
# Runs baseline and controlled variation experiments
# Automated input for stock symbol and date range

set -euo pipefail

ROOT_DIR="$(cd "$(dirname "${BASH_SOURCE[0]}")/.." && pwd)"
RUNS_DIR="${ROOT_DIR}/runs"
LOG_DIR="${ROOT_DIR}/logs"

mkdir -p "${RUNS_DIR}" "${LOG_DIR}"

export PYTHONHASHSEED=${PYTHONHASHSEED:-42}
SEED=${SEED:-42}

echo "Timestamp: $(date -Is)"
echo "Python: $(python -V 2>&1 || true)"
echo "Pip: $(pip -V 2>&1 || true)"
echo "OS: $(uname -a)"
echo "Git commit (repo/): $(git -C "${ROOT_DIR}/repo" rev-parse HEAD 2>/dev/null || echo 'n/a')"
echo "CUDA: ${CUDA_VISIBLE_DEVICES:-unset}"

# --- Baseline run ---

BASELINE_OUT="${RUNS_DIR}/baseline"
mkdir -p "${BASELINE_OUT}"
echo "=== BASELINE ===" | tee -a "${LOG_DIR}/commands.txt"

BASELINE_CMD=(python main.py)
printf 'CMD: %q ' "${BASELINE_CMD[@]}" | tee -a "${LOG_DIR}/commands.txt"

# Run baseline with inputs for AAPL and date range
{ printf "AAPL\n2025-05-28\n2025-06-30\n"; } | "${BASELINE_CMD[@]}" 2>&1 | tee "${BASELINE_OUT}/stdout.log"

# --- Controlled Variation run ---

VAR_OUT="${RUNS_DIR}/variation"
mkdir -p "${VAR_OUT}"
echo "=== VARIATION ===" | tee -a "${LOG_DIR}/commands.txt"

VARIATION_CMD=(python main.py)
printf 'CMD: %q ' "${VARIATION_CMD[@]}" | tee -a "${LOG_DIR}/commands.txt"

# Run variation with input for TSLA and same date range
{ printf "TSLA\n2025-05-28\n2025-06-30\n"; } | "${VARIATION_CMD[@]}" 2>&1 | tee "${VAR_OUT}/stdout.log"

echo "All done. Outputs in: ${RUNS_DIR} | Logs in: ${LOG_DIR}"
```
